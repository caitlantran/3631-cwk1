Assume the initial position of the end-effector is p = (1, 2, 3). The goal is for the end-effector to pass through a via-point (5, 10, 12) at t = 1s and reach the goal at (10, 8, 7) by t = 5s.

1. Generate a 3D trajectory for the end-effector using two second-order polynomials, one for each dimension (planning in Cartesian space).
2. Plot the generated trajectories in the form of t-x, t-y, and t-z (one figure for each dimension).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Second-order polynomials trajectory
# x(t) = a_x * t^2 + b_x * t + c_x
# y(t) = a_y * t^2 + b_y * t + c_y
# z(t) = a_z * t^2 + b_z * t + c_z

# define the time and position of start point, via point and end-point
start_point, via_point, end_point = np.array([1, 2, 3]), np.array([5, 10, 12]), np.array([10, 8, 7])
points = np.array([
    start_point,
    via_point,
    end_point
])
start_time, via_time, end_time = 0.0, 1.0, 5.0

# TODO: implement your code here. When you plot your trajectories, remember to display the start-, via-, and end-point.
def compute_coefficients(t0, t1, t2, p0, p1, p2):
    A = np.array([
        [t0**2, t0, 1],
        [t1**2, t1, 1],
        [t2**2, t2, 1]
    ])
    coeffs_x = np.linalg.solve(A, [p0[0], p1[0], p2[0]])
    coeffs_y = np.linalg.solve(A, [p0[1], p1[1], p2[1]])
    coeffs_z = np.linalg.solve(A, [p0[2], p1[2], p2[2]])
    return coeffs_x, coeffs_y, coeffs_z

coeffs_x, coeffs_y, coeffs_z = compute_coefficients(start_time, via_time, end_time, start_point, via_point, end_point)

t_values = np.linspace(start_time, end_time, 100)
x_values = coeffs_x[0] * t_values**2 + coeffs_x[1] * t_values + coeffs_x[2]
y_values = coeffs_y[0] * t_values**2 + coeffs_y[1] * t_values + coeffs_y[2]
z_values = coeffs_z[0] * t_values**2 + coeffs_z[1] * t_values + coeffs_z[2]

plt.figure()
plt.plot(t_values, x_values, label='x(t)')
plt.scatter([start_time, via_time, end_time], [start_point[0], via_point[0], end_point[0]], color='red', zorder=3)
plt.xlabel('Time (s)')
plt.ylabel('X Position')
plt.legend()
plt.title('Time vs X Position')
plt.show()

plt.figure()
plt.plot(t_values, y_values, label='y(t)')
plt.scatter([start_time, via_time, end_time], [start_point[1], via_point[1], end_point[1]], color='red', zorder=3)
plt.xlabel('Time (s)')
plt.ylabel('Y Position')
plt.legend()
plt.title('Time vs Y Position')
plt.show()

plt.figure()
plt.plot(t_values, z_values, label='z(t)')
plt.scatter([start_time, via_time, end_time], [start_point[2], via_point[2], end_point[2]], color='red', zorder=3)
plt.xlabel('Time (s)')
plt.ylabel('Z Position')
plt.legend()
plt.title('Time vs Z Position')
plt.show()



```
